```python
def delta(arr, map_function=None, use_map=False):
    if map_function is not None:
        if use_map:
            arr = map(lambda x: map_function(x), arr)
        else:
            arr = map_function(arr)
        
    return np.array(arr[1:] - arr[:-1])
```

```python
log_delta_X = delta(X_SPOT, np.log)
log_delta_Y = delta(Y_SPOT, np.log)

rho = np.corrcoef(log_delta_X, log_delta_Y)

sigma_X = np.std(log_delta_X)
sigma_Y = np.std(log_delta_Y)
```

```python
N = 4
r = 0.01
```

```python
n_partitions = 252
m = 10000

dt = N

spot_X = X_SPOT[-1]
spot_Y = Y_SPOT[-1]

rdt = r * dt
sigmadt_X = sigma_X * math.sqrt(dt)
sigmadt_Y = sigma_Y * math.sqrt(dt)
```

```python
partitions_to_check = dict([((0.25 * i * 252) / N - 1, True) for i in range(1, N + 1)])

q_spot_X = [0.75 * spot_X, 1.25 * spot_X]
q_spot_Y = [0.75 * spot_Y, 1.25 * spot_Y]
```

```python
values = []
for _ in range(m):
    x = spot_X
    y = spot_Y
    
    dW = np.random.multivariate_normal([0, 0], rho, n_partitions)
    
    accept = True
    
    for i in range(n_partitions):
        x += x * rdt + sigmadt_X * dW[i][0]
        y += y * rdt + sigmadt_Y * dW[i][1]
        
        if i in partitions_to_check:
            if not (q_spot_X[0] <= x <= q_spot_X[1]):
                accept = False
                break
            
            if not (q_spot_Y[0] <= y <= q_spot_Y[1]):
                accept = False
                break

    if accept:
        values.append([x, y])
```
